## Chapter 04 - 1. DataFrame Group by 이해하기 

```python
import pandas as pd
import numpy as np
```

**group by**

- 아래의 세 단계 적용, 데이터를 그룹화(groupping) (SQL의 group by와 개념적으로 동일, 사용법은 유사)
    - 데이터 분할
    - operation 적용
    - 데이터 병합

```python
df = pd.read_csv('./train.csv')
df.head()
```

**GroupBy groups 속성**

- 각 그룹과 그룹에 속한 index를 dict 형태로 표현

```python
class_group = df.groupby('Pclass')
class_group
```

```python
# key에 해당하는 인덱스(리스트) 값으로 나타남.
class_group.groups
```

```python
gender_group = df.groupby('Sex')
gender_group.groups
```

**groupping 함수**
- 그룹 데이터에 적용 가능한 통계 함수(NaN은 제외하여 연산)
- count: 데이터 개수
- sum: 데이터 합
- mean, std, var: 평균, 표준편차, 분산
- min, max: 최소, 최대값

```python
class_group.count()
```

```python
class_group.std()
```

```python
# 생존율 계산
class_group.mean()['Survived']
```

- 성별에 따른 생존율 구해보기

```python
df.groupby('Sex').mean()['Survived']
```

**복수 columns로 groupping하기**
- groupby에 column 리스트를 전달
- 통계함수를 적용한 결과는 multiindex를 갖는 dataframe

- 클래스와 성별에 따라 생존률 구해보기

```python
df.groupby(['Pclass', 'Sex']).mean()
```

**index를 이용한 group by**

- index가 있는 경우, groupby함수에 level 사용 가능
    - level은 index의 depth를 의미하며, 가장 왼쪽부터 0부터 증가
    

- **set_index** 함수
    - column 데이터를 index 레벨로 변경
- **reset_index** 함수
    - 인덱스 초기화

```python
df.groupby(['Pclass', 'Sex']).mean().loc[(2, 'female')]
```

```python
df.head()
```

```python
df.set_index('Pclass')
```

```python
df.set_index(['Pclass', 'Sex'])
```

```python
df.set_index(['Pclass', 'Sex']).reset_index()
```

```python
# 위의 경우, Age로 인덱싱 하라는 뜻.

df.set_index('Age').groupby(level=0).mean()
```

### 나이대별로 생존율 구하기

```python
import math
def age_categorize(age):
    if math.isnan(age):
        return -1
    return math.floor(age/10) * 10
    
```

```python
df.set_index('Age').groupby(age_categorize).mean()['Survived']
```

### Multiindex를 이용한 grouping

```python
df.head()
```

```python
df.set_index(['Pclass', 'Sex']).groupby(level=[0, 1]).mean()
```

### aggregate Function 
- groupby 결과에 집계함수를 적용하여, 그룹별 데이터 확인 가능

```python
df.set_index(['Pclass', 'Sex']).groupby(level=[0, 1]).aggregate([np.mean, np.sum, np.max])
```
