## Chapter 04 - 2. transform 함수의 이해 및 활용

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./train.csv')
df.head()
```

### transform function
- groupby 후, transform 함수를 사용하면 **기존의 index 유지한 상태**로 통계함수를 적용
- 전체 데이터의 집계가 아닌, 각 그룹에서의 집계를 계산
- 따라서 새로 생성된 데이터를 원본 dataframe과 합치기 쉬움 

```python
df.groupby('Pclass').mean()
```

```python
df.groupby('Pclass').transform(np.mean)
```

```python
df['Age2'] = df.groupby('Pclass').transform(np.mean)['Age']
df.head()
```

```python
df.groupby(['Pclass', 'Sex']).mean()
```

```python
df['Age3'] = df.groupby(['Pclass', 'Sex']).transform(np.mean)['Age']
df.head()
```

## Chapter 04 - 3. Pivot, Pivot_table 함수의 이해 및 활용

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    '지역' : ['서울', '서울', '서울', '경기', '경기', '부신', '서울', '서울', '부산', '경기', '경기', '경기'],
    '요일' : ['월요일', '화요일', '수요일', '월요일', '화요일', '월요일', '목요일', '금요일', '화요일', '수요일', '목요일', '금요일'],
    '강수량': [100, 80, 1000, 200, 200, 100, 50, 100, 200, 100, 50, 100],
    '강수확률' : [80, 70, 90, 10, 20, 30, 50, 90, 20, 80, 50, 10]
})
df
```

### pivot
- dataframe의 형태로 변경
- 인덱스, 컬럼, 데이터로 사용할 컬럼 명시

```python
df.pivot('지역', '요일')
```

```python
# 2번째 값: Values
df.pivot('요일', '지역', '강수량')
```

**pivot_table**
- 기능적으로 pivot과 동일
- pivot과의 차이점
    - 중복되는 모호한 값이 있을 경우, aggregation 함수 사용하여 값을 채움

```python
df = pd.DataFrame({
    '지역' : ['서울', '서울', '서울', '경기', '경기', '부산', '서울', '서울', '부산', '경기', '경기', '경기'],
    '요일' : ['월요일', '월요일', '수요일', '월요일', '화요일', '월요일', '목요일', '금요일', '화요일', '수요일', '목요일', '금요일'],
    '강수량': [100, 80, 1000, 200, 200, 100, 50, 100, 200, 100, 50, 100],
    '강수확률' : [80, 70, 90, 10, 20, 30, 50, 90, 20, 80, 50, 10]
})
df
```

```python
# 중복 허용 후 추출
df.pivot('요일', '지역')
```

```python
# 서울의 경우, 1월요일 강수량이 100+80 / 2 의 값이 도출된 것을 확인할 수 있음.

df.pivot_table(df, index='요일', columns='지역', aggfunc=np.mean)
```

## Chapter 04 - 4. stack, unstack 함수의 이해 및 활용

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    '지역' : ['서울', '서울', '서울', '경기', '경기', '부신', '서울', '서울', '부산', '경기', '경기', '경기'],
    '요일' : ['월요일', '화요일', '수요일', '월요일', '화요일', '월요일', '목요일', '금요일', '화요일', '수요일', '목요일', '금요일'],
    '강수량': [100, 80, 1000, 200, 200, 100, 50, 100, 200, 100, 50, 100],
    '강수확률' : [80, 70, 90, 10, 20, 30, 50, 90, 20, 80, 50, 10]
})
df
```

### stack & unstack
- stack: 칼럼 레벨에서, 인덱스 레벨로 dataframe 변경
    - 즉, 데이터를 쌓아올리는 개념으로 이해하면 쉬움.
- unstack: 인덱스 레벨에서 컬럼 레벨로 dataframe 변경
    - unstack의 반대 operation
- 둘은 역의 관계에 있음.

```python
new_df = df.set_index(['지역', '요일'])
new_df.head()
```

```python
# level = 인덱스 레벨
# 0 -> 지역 index를 컬럼으로

new_df.unstack(0)
```

```python
new_df.unstack(1)
```

```python
new_df.unstack(0).stack(0)
```

```python
new_df.unstack(0).stack(1)
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
pd.concat([df1, df2], axis=1)
```

```python
pd.concat([df1, df3], axis=1)
```

## Chapter 05 -2. Merge & join function, 데이터 프레임 병합

```python
import numpy as np
import pandas as pd
```

### dataframe merge
- SQL join처럼, 특정한 column을 기준으로 병합
    - join방식: how 파라미터 통해 명시
        - inner: 기본값, 일치하는 값 있는 경우.
        - left: left outer join
        - right: right outer join
        - outer: full outer join

```python
customer = pd.DataFrame({'c_id' : np.arange(6), 
                        'name': ['Jane', 'Yong', 'Roony', 'Park', 'Su', 'Harry'],
                        'age' : [40, 20, 21, 30, 31, 18]})
customer
```

```python
orders = pd.DataFrame({'c_id' : [1, 1, 2, 2, 2, 3, 3, 1, 4, 9],
                      'item': ['Snack', 'Phone', 'Monitor', 'Noodle', 'Earphone', 'Glass', 'Pen', 'Note', 'Sunblock', 'USB'],
                      'quantity': [1, 2, 1, 1, 3, 2, 2, 3, 2, 1]})
orders.head()
```

- on
    - join 대상이 되는 column 명시

```python
# default = 'inner'
pd.merge(customer, orders, on='c_id')
```

```python
pd.merge(customer, orders, on='c_id', how='left')
```

```python
pd.merge(customer, orders, on='c_id', how='right')
```

```python
pd.merge(customer, orders, on='c_id', how='outer')
```

- index 기준으로 join 

```python
cust1 = customer.set_index('c_id')
order1 = orders.set_index('c_id')
```

```python
cust1
```

```python
pd.merge(cust1, order1, left_index=True, right_index=True)
```

### Practice
- 가장 많이 팔린 아이템은?
- Yong이 가장 많이 구매한 아이템은?

```python
pd.merge(customer, orders, on='c_id').groupby('item').sum().sort_values(by='quantity', ascending=False)
```

```python
pd.merge(customer, orders, on='c_id').groupby(['name', 'item']).sum().sort_values(by=['name', 'quantity'], ascending=False)
```

```python
pd.merge(customer, orders, on='c_id').groupby(['name', 'item']).sum().loc['Yong']
```

```python
pd.merge(customer, orders, on='c_id').groupby(['name', 'item']).sum().loc['Yong', 'quantity']
```

### join
- 내부적으로 pandas.merge 사용
- 기본적으로 index를 사용하여 left join

```python
cust1.join(order1, how='inner')
```
